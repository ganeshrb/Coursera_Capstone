```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
```

```python
mu = 50
sigma = 10
x = mu + sigma * np.random.randn(500)

#plt.figure(figsize=(4, 4))
plt.hist(x, bins=50, density=True, color ='g', edgecolor='k', alpha=0.7)

# Plot the PDF.
xmin, xmax = plt.xlim() # it will find xmin and xmax from the plot of histogtam
t = np.linspace(xmin, xmax, 500)# 1000 equally spaced values between xmin and xmax
y = norm.pdf(t, mu, sigma) # probability density function of normal distribution
plt.plot(t, y, 'r', linewidth=2)
plt.xlabel('Smarts', fontsize=14)
plt.ylabel('Probability', fontsize=14)
plt.title('Histogram of IQ', fontsize=20)
plt.show()
```
